## Metodo de Euler

$$
y_{+1}=y_{i}+hf(x_{i},y_{i})
$$


$$
x_{i+1}=x_{i}+h,   i=0,1,2,...
$$


$$
E=\frac{h^2}{2!}y''(z)=\mathscr{O}(h^2)
$$


## Euler Mejorado (Heun)

$$
\overline{y_{i+1}}=y_i+hf(x_i, y_i)
$$


$$
y_{i+1}=y_i+\frac{h}{2}(f(x_i, y_i)+f(x_{i+1},\overline{y_{i+1}}))
$$


$$
x_{i+1}=x_i+h   i=1,2,...
$$


$$
E=\frac{h^3}{3!}y''(z)=\mathscr{O}(h^3)
$$


## Runge-Kutta (Segundo Orden)

$$
k_1=hf(x_i,y_i)
$$


$$
k_2=hf(x_i+h,y_i+k_1)
$$


$$
y_{i+1}=y_i+\frac{1}{2}(k_1+k_2)
$$


$$
x_{i+1]=x_i+h   i=0,1,2,...
$$


$$
E=\frac{h^3}{3!}y''(z)=\mathscr{O}(h^3)
$$


## Runge-Kutta (Cuarto Orden)

$$
k_1=hf(x_i,y_i)
$$


$$
k_2=hf(x_i+\frac{h}{2},y_i+\frac{k_1}{2})
$$


$$
k_3=hf(x_i+\frac{h}{2},y_i+\frac{k_2}{2})
$$


$$
k_4=hf(x_i+h,y_i+k_3)
$$


$$
y_{i+1}=y_i+\frac{1}{6}(x_i+2k_2+2k_3+k_4)
$$


$$
x_{i+1}=x_i+h    i=0,1,2,...
$$


$$
E=\mathscr{O}(h^5)
$$

```python
import pylab as pl
import numpy as np

def euler(f,x,y,h,m):
    u=np.zeros([m,2])
    for i in range(m):
        y=y+h*f(x,y)
        x=x+h
        u[i,0]=x
        u[i,1]=y
    return u

def f(x,y):
    return y-x**2+x+1

u=euler(f,0,1,0.1,20)
pl.plot(u[:,0],u[:,1],'or', label='Metodo de Euler')

def y(x):
    return np.exp(x)+x**2+x
x=pl.arange(0,2.1,0.1)
pl.plot(x,y(x), label='Solucion Exacta')
pl.grid(True)
pl.legend()
pl.show()
```

## Método mejorado de Euler (Heun)

```python
import numpy as np
import pylab as pl

def heun(f,x,y,h,m):
    u=np.zeros([m,2],dtype=float)
    for i in range(m):
        yn=y+h*f(x,y)
        y=y+h/2*(f(x,y)+f(x+h,yn))
        x=x+h
        u[i,0]=x
        u[i,1]=y
    return u

def f(x,y):
    return y-x**2+x+1

u=heun(f,0,1,0.1,20)
pl.plot(u[:,0],u[:,1],'or', label='Metodo de Heun')

def y(x):
    return np.exp(x)+x**2+x
x=pl.arange(0,2.1,0.1)
pl.plot(x,y(x), label='Solucion Exacta')
pl.grid(True)
pl.legend()
pl.show()
```

## Runge-Kutta (Segundo Orden)

```python
import numpy as np
import pylab as pl

def rk2(f,x,y,h,m):
    u=np.zeros([m,2],dtype=float)
    for i in range(m):
        k1=h*f(x,y)
        k2=h*f(x+h,y+k1)
        y=y+0.5*(k1+k2)
        x=x+h
        u[i,0]=x
        u[i,1]=y
    return u

def f(x,y):
    return y-x**2+x+1

u=heun(f,0,1,0.1,20)
pl.plot(u[:,0],u[:,1],'or', label='Metodo de Runge-Kutta')

def y(x):
    return np.exp(x)+x**2+x
x=pl.arange(0,2.1,0.1)
pl.plot(x,y(x), label='Solucion Exacta')
pl.grid(True)
pl.legend()
pl.show()
```

## Runge-Kutta (Cuarto Orden)

```python
import numpy as np
import pylab as pl

def rk4(f,x,y,h,m):
    u=np.zeros([m,2],dtype=float)
    for i in range(m):
        k1=h*f(x,y)
        k2=h*f(x+h/2,y+k1/2)
        k3=h*f(x+h/2,y+k2/2)
        k4=h*f(x+h,y+k3)
        y=y+1/6*(k1+2*k2+2*k3+k4)
        x=x+h
        u[i,0]=x
        u[i,1]=y
    return u

def f(x,y):
    return y-x**2+x+1

u=rk4(f,0,1,0.1,20)
pl.plot(u[:,0],u[:,1],'or', label='Metodo de Runge-Kutta')

def y(x):
    return np.exp(x)+x**2+x
x=pl.arange(0,2.1,0.1)
pl.plot(x,y(x), label='Solucion Exacta')
pl.grid(True)
pl.legend()
pl.show()
```
