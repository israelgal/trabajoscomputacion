```python
#Autor: Jose Israel Galindo Rodriguez



import math 
from sympy import *
import numpy as np
x=symbols('x')

#funcion a integrar


f = x**3 + 2*x**2 




# definicion de factorial 
def factorial(n):
    M=1
    F=1
    while M<float(n):
        M=M+1
        F=F*M
    return (F)
################ funcion polinomio de legengre #######
def polilegengre(x,n):
    fun=(x**2-1)**float(n)
    derivada=diff(fun,x,n)
    polinomio=(1/(2**float(n)*factorial(n)))*derivada
    return (polinomio)

############### derivada de polinomio de legengre #######

def derpolinomio(x,n):
    derivadadepolinomio=diff(polilegengre(x,n),x,1)
    polinomio2=derivadadepolinomio
    return derivadadepolinomio

############# funcion para los pesos wi ################
def pesow(x,n):
    wi=2/((1-x**2)*(derpolinomio(x,n))**2)
    return wi

n = int(input("Ingrese n: "))


a=input('punto inicial de tu integral')
b=input('punto final de integral')


# cambio de variable de la funcion y rango de la integral 

rango1=(float(b)-float(a))/2
rango12=(float(b)-float(a))*x/2
rango2=(float(a)+float(b))/2



# integral
cambiovariable=f.subs(x,rango12+rango2)

# puntos xi

xi = np.arange(0.0,float(n))


for i in range(0,n):
  print ("Ingrese el grado de tu polinomio",i+1,": ")
  xi[i] = float(input())

integral=1.0
casi=0

for i in range(0,n):
    integral=(pesow(x,n).subs(x,xi[i]))*cambiovariable.subs(x,xi[i])
    casi = casi + integral
              

integral3=rango1*casi 

print(integral3)              
              
    

      
    
```
